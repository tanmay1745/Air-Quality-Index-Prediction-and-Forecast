```python
from datetime import datetime, timedelta
from meteostat import Point, Hourly
from geopy.geocoders import Nominatim
from tqdm import tqdm
import os

# ---------- User params ----------
AQI_DF_PATH = "Cleaned_Pollutant_Data.csv"          # path to your cleaned dataframe (or you already have it in memory)
STATION_COORDS_PATH = "station_coords.csv" # Should have StationId, Latitude, Longitude (optional)
LOCAL_TZ = "Asia/Kolkata"  # change if your data timezone is different
CACHE_DIR = "weather_cache"  # caching to avoid repeated downloads
# ----------------------------------

os.makedirs(CACHE_DIR, exist_ok=True)

# 1) load the main df (or assume you have it as `df`)
df = pd.read_csv(AQI_DF_PATH, parse_dates=['Datetime'])
# If you already have df in memory, skip loading and make sure Datetime is datetime dtype
# df = your_dataframe  # << uncomment/replace if needed

# Ensure correct dtypes and round datetimes to hour (important for joining)
df['Datetime'] = pd.to_datetime(df['Datetime'])
# Round down to hour to align with hourly weather
df['Datetime_hour'] = df['Datetime'].dt.floor('H')

# 2) get station -> coords mapping
if os.path.exists(STATION_COORDS_PATH):
    station_coords = pd.read_csv(STATION_COORDS_PATH, dtype={'StationId': str})
    assert {'StationId','Latitude','Longitude'}.issubset(station_coords.columns), \
        "station_coords.csv must contain StationId, Latitude, Longitude"
else:
    # Build station_coords by geocoding StationName (fallback)
    geolocator = Nominatim(user_agent="aqi_station_geocoder")
    unique_stations = df[['StationId','StationName','City']].drop_duplicates(subset='StationId')
    rows = []
    print("No station_coords.csv found — attempting geocoding (may be slow).")
    for _, r in tqdm(unique_stations.iterrows(), total=len(unique_stations)):
        sid = str(r['StationId'])
        name = r.get('StationName') or ""
        city = r.get('City') or ""
        q = f"{name}, {city}"
        try:
            loc = geolocator.geocode(q, timeout=10)
            if loc:
                rows.append({'StationId': sid, 'Latitude': loc.latitude, 'Longitude': loc.longitude})
            else:
                rows.append({'StationId': sid, 'Latitude': np.nan, 'Longitude': np.nan})
        except Exception as e:
            rows.append({'StationId': sid, 'Latitude': np.nan, 'Longitude': np.nan})
    station_coords = pd.DataFrame(rows)
    station_coords.to_csv(STATION_COORDS_PATH, index=False)
    print(f"Saved station_coords to {STATION_COORDS_PATH}. Manually fix any NaNs if needed.")

# Merge station coords into df for easier grouping
df = df.merge(station_coords, on='StationId', how='left')

# If any stations still missing coords: report them
missing_coords = station_coords[station_coords[['Latitude','Longitude']].isnull().any(axis=1)]
if len(missing_coords) > 0:
    print("WARNING: Some stations are missing coordinates. Fix station_coords.csv for those StationIds:")
    print(missing_coords)

# 3) helper to fetch hourly weather with caching
def fetch_weather_for_point(lat, lon, start, end, cache_path):
    """
    Uses Meteostat Hourly API for the given lat/lon and date range.
    Caches results by cache_path (csv).
    """
    if os.path.exists(cache_path):
        try:
            w = pd.read_csv(cache_path, parse_dates=['time'])
            return w
        except:
            pass

    pt = Point(lat, lon)
    # meteostat expects timezone-aware datetimes or naive UTC; we'll pass naive UTC-ish ranges
    # fetch hourly
    hourly = Hourly(pt, start, end)
    try:
        w = hourly.fetch()
    except Exception as e:
        print(f"Failed to fetch meteostat for {lat},{lon} => {e}")
        return pd.DataFrame()
    # Reset index (time) as column and keep only useful columns
    w = w.reset_index().rename(columns={'time': 'Datetime_hour'})
    # Save cache
    w.to_csv(cache_path, index=False)
    return w

# 4) loop over stations and fetch / merge
weather_frames = []
stations = df[['StationId','Latitude','Longitude']].drop_duplicates(subset='StationId')

for _, s in tqdm(stations.iterrows(), total=len(stations)):
    sid = s['StationId']
    lat = s['Latitude']
    lon = s['Longitude']
    if pd.isna(lat) or pd.isna(lon):
        print(f"Skipping Station {sid} (no coords).")
        continue

    # restrict date range to only what we need for that station to reduce API calls
    df_s = df[df['StationId'] == sid]
    start = df_s['Datetime_hour'].min().floor('D')  # start of earliest day
    end = df_s['Datetime_hour'].max().ceil('D') + pd.Timedelta(days=1)  # inclusive

    # prepare cache filename
    cache_file = os.path.join(CACHE_DIR, f"weather_sid_{sid}_{start.date()}_{end.date()}.csv")

    w = fetch_weather_for_point(lat, lon, start, end, cache_file)
    if w.empty:
        continue

    # Meteostat returns columns like: temp, dwpt, rh, prcp, snow, wdir, wspd, wpgt, pres
    # Keep only columns we want; rename time column to Datetime_hour which we already did
    # Convert timezone if needed (meteostat returns tz-aware times) -> make naive and convert to same timezone as df
    if 'Datetime_hour' in w.columns:
        # Make sure it's datetime and round to hour
        w['Datetime_hour'] = pd.to_datetime(w['Datetime_hour'])
        w['Datetime_hour'] = w['Datetime_hour'].dt.floor('h')

    # add StationId column so we can concat all and merge back
    w['StationId'] = sid
    # optionally reduce to useful columns
    keep_cols = ['StationId', 'Datetime_hour', 'temp', 'dwpt', 'rh', 'prcp', 'snow', 'wdir', 'wspd', 'pres']
    # Keep only intersection
    keep_cols = [c for c in keep_cols if c in w.columns]
    w = w[keep_cols]
    weather_frames.append(w)

# concat all station weather data
if len(weather_frames) == 0:
    raise RuntimeError("No weather data fetched. Check coords and internet access.")
weather_all = pd.concat(weather_frames, ignore_index=True)

# 5) Merge back into original df on StationId + Datetime_hour
# If there are duplicates for the same station+hour (unlikely), aggregate by mean
weather_all = weather_all.groupby(['StationId','Datetime_hour'], as_index=False).mean()

# merge (left join keeps all AQI rows)
df_merged = df.merge(weather_all, on=['StationId','Datetime_hour'], how='left', suffixes=('','_weather'))

# 6) optional: forward/backward fill short gaps in weather (per station)
# For each station, fill small gaps (<=2 hours) with interpolation
def fill_small_gaps(group, cols, max_gap_hours=2):
    group = group.sort_values('Datetime_hour')
    for c in cols:
        group[c] = group[c].interpolate(limit=max_gap_hours, limit_direction='both')
    return group

weather_cols = [c for c in weather_all.columns if c not in ['StationId','Datetime_hour']]
df_merged = df_merged.groupby('StationId').apply(lambda g: fill_small_gaps(g, weather_cols)).reset_index(drop=True)

# 7) Save result
df_merged.to_parquet("aqi_with_weather.parquet", index=False)   # efficient for large frames
print("Saved merged file to aqi_with_weather.parquet")
```

### Weather Columns & Units

- temp → °C
- dwpt → °C
- wdir → degrees (0–360)
- wspd → m/s
- pres → hPa

```python
import numpy as np

def compute_rh(temp, dewpoint):
    a, b = 17.625, 243.04
    alpha = (a * dewpoint) / (b + dewpoint)
    beta = (a * temp) / (b + temp)
    rh = 100 * np.exp(alpha - beta)
    return rh

weather_all["RelativeHumidity"] = compute_rh(weather_all["Temperature"], weather_all["DewPoint"])
```

```python
def fix_weather_nan_only(df):

    df = df.sort_values("Datetime").copy()

    # Set datetime as index for "time" interpolation
    df = df.set_index("Datetime")

    # -------- 1. Temperature & DewPoint (interpolate only NaNs) --------
    for col in ["Temperature", "DewPoint"]:
        if col in df.columns:
            mask = df[col].isna()
            interp = df[col].interpolate(method="time", limit=3, limit_direction="both")
            df.loc[mask, col] = interp[mask]

    # -------- 2. Relative Humidity (compute only where NaN) --------
    if all(c in df.columns for c in ["RelativeHumidity", "Temperature", "DewPoint"]):
        rh_mask = df["RelativeHumidity"].isna()

        a, b = 17.625, 243.04
        alpha = (a * df["DewPoint"]) / (b + df["DewPoint"])
        beta  = (a * df["Temperature"]) / (b + df["Temperature"])
        rh_new = 100 * np.exp(alpha - beta)

        df.loc[rh_mask, "RelativeHumidity"] = rh_new[rh_mask]

    # -------- 3. Wind: direction = NaN when speed = 0 --------
    if "WindSpeed" in df.columns and "WindDirection" in df.columns:
        zero_mask = df["WindSpeed"] == 0
        df.loc[zero_mask, "WindDirection"] = np.nan

    # -------- 4. WindSpeed interpolation only on NaNs --------
    if "WindSpeed" in df.columns:
        mask = df["WindSpeed"].isna()
        interp = df["WindSpeed"].interpolate(limit=3, limit_direction="both")
        df.loc[mask, "WindSpeed"] = interp[mask]

    # -------- 5. WindDirection circular interpolation only on NaNs --------
    if "WindDirection" in df.columns:
        mask = df["WindDirection"].isna()

        rad = np.deg2rad(df["WindDirection"])
        sin = np.sin(rad)
        cos = np.cos(rad)

        sin_interp = sin.interpolate(limit=3, limit_direction="both")
        cos_interp = cos.interpolate(limit=3, limit_direction="both")

        new_dir = np.rad2deg(np.arctan2(sin_interp, cos_interp))
        new_dir = (new_dir + 360) % 360

        df.loc[mask, "WindDirection"] = new_dir[mask]

    # -------- 6. Pressure fill (forward + backward) only for NaN --------
    if "Pressure" in df.columns:
        mask = df["Pressure"].isna()
        ffilled = df["Pressure"].fillna(method="ffill").fillna(method="bfill")
        df.loc[mask, "Pressure"] = ffilled[mask]

    # finally reset index back to normal
    df = df.reset_index()

    return df
```

```python
# Step 1: Drop unwanted columns
df_final = df_final.drop(columns=["Unnamed: 0.1", "Unnamed: 0", "Datetime_hour", "StationName_y", "City_y"])
```

```python
df_final = df_final.sort_values(["StationId", "Datetime"]).reset_index(drop=True)
```

```python
df_final.duplicated().sum()
```

```python
# Convert nullable float columns (Float64) to normal float dtype (float64)
# Identify columns with pandas nullable "Float64" type
nullable_float_cols = df_final.columns[df_final.dtypes == "Float64"].tolist()
print("Nullable Float64 columns:", nullable_float_cols)

# Convert them to float64 (standard)
for i in nullable_float_cols:
    df_final[i] = df_final[i].astype('float64')

df_final.info()
```

```python
# Time-based feature engineering

df_final["DayOfWeek"] = df_final["Datetime"].dt.dayofweek
df_final["IsWeekend"] = df_final["DayOfWeek"].isin([5, 6]).astype(int)
df_final["Month"] = df_final["Datetime"].dt.month
df_final["Quarter"] = df_final["Datetime"].dt.quarter
df_final["DayOfYear"] = df_final["Datetime"].dt.dayofyear
df_final["Year"] = df_final["Datetime"].dt.year

# Season (Winter / Summer / Monsoon / Post-Monsoon)
def get_season(m):
    if m in [12, 1, 2]:
        return "Winter"
    elif m in [3, 4, 5]:
        return "Summer"
    elif m in [6, 7, 8, 9]:
        return "Monsoon"
    else:
        return "PostMonsoon"

df_final["Season"] = df_final["Month"].apply(get_season)

df_final.head()
```

```python
df_final.info()
```

```python
# Drop columns we don't need
df_final = df_final.drop(columns=["AQI_Bucket", "StationName_x", "City_x"], axis=1)
```

```python
# Step 6: Encoding categorical columns
from sklearn.preprocessing import LabelEncoder

# Label Encode StationId
le = LabelEncoder()
df_final["StationId_enc"] = le.fit_transform(df_final["StationId"])

# One-hot encode Day and Season
df_final = pd.get_dummies(df_final, columns=["Day", "Season"], drop_first=True)

df_final.shape
```

```python
df_final.info()
```

```python
# Convert bool dummies to int
bool_cols = df_final.select_dtypes(include='bool').columns
df_final[bool_cols] = df_final[bool_cols].astype(int)

# Drop latitude & longitude (not required, model learns location from StationId_enc)
df_final = df_final.drop(columns=["Latitude", "Longitude"])
```

```python
df_final.info()
```

```python
import numpy as np
import pandas as pd

# Ensure datetime sorted properly for lag/rolling
df_final = df_final.sort_values("Datetime").reset_index(drop=True)

# -----------------------------
# 1. LAG FEATURES
# -----------------------------
df_final["PM25_lag1"]  = df_final.groupby("StationId")["PM2.5"].shift(1)
df_final["PM25_lag3"]  = df_final.groupby("StationId")["PM2.5"].shift(3)
df_final["PM25_lag6"]  = df_final.groupby("StationId")["PM2.5"].shift(6)
df_final["PM25_lag24"] = df_final.groupby("StationId")["PM2.5"].shift(24)

df_final["NO2_lag1"]   = df_final.groupby("StationId")["NO2"].shift(1)
df_final["O3_lag1"]    = df_final.groupby("StationId")["O3"].shift(1)

# -----------------------------
# 2. ROLLING WINDOW FEATURES
# -----------------------------
df_final["PM25_roll3"] = (
    df_final.groupby("StationId")["PM2.5"]
    .rolling(3)
    .mean()
    .reset_index(level=0, drop=True)
)

df_final["PM25_roll6"] = (
    df_final.groupby("StationId")["PM2.5"]
    .rolling(6)
    .mean()
    .reset_index(level=0, drop=True)
)

df_final["PM25_roll12"] = (
    df_final.groupby("StationId")["PM2.5"]
    .rolling(12)
    .mean()
    .reset_index(level=0, drop=True)
)

df_final["PM25_roll24"] = (
    df_final.groupby("StationId")["PM2.5"]
    .rolling(24)
    .mean()
    .reset_index(level=0, drop=True)
)

# -----------------------------
# 3. CYCLIC TIME ENCODING
# -----------------------------
df_final["Hour_sin"]   = np.sin(2 * np.pi * df_final["Hour"] / 24)
df_final["Hour_cos"]   = np.cos(2 * np.pi * df_final["Hour"] / 24)

df_final["Month_sin"]  = np.sin(2 * np.pi * df_final["Month"] / 12)
df_final["Month_cos"]  = np.cos(2 * np.pi * df_final["Month"] / 12)

# -----------------------------
# 4. POLLUTION LOAD / RATIOS
# -----------------------------
df_final["Pollution_Load"] = (
    df_final["PM2.5"] + df_final.get("PM10", 0) + df_final["NO2"]
)

df_final["PM_Ratio"] = df_final["PM2.5"] / (df_final.get("PM10", 0) + 1)

# -----------------------------
# 5. INTERACTION FEATURES
# -----------------------------
df_final["Temp_Humidity_Interaction"] = (
    df_final["Temperature"] * df_final["RelativeHumidity"]
)

df_final["Wind_Inv"] = 1 / (df_final["WindSpeed"] + 0.1)

# Done
print("Advanced feature engineering completed.")
print(df_final.shape)
```
